```python
import pandas as pd
import os
```

```python
models = next(os.walk('.'))[1]
print(models)
```

```python
biases = ['acquiescence', 'negative_wording', 'odd_even', 'opinion_float', 'response_order']
# rename "sample" files
sample_models = ['llama2-70b-ift', 'gpt-3.5-turbo']
for model in sample_models:
    for bias in biases:
        filename = f"{bias}-sample.pickle"
        filepath = os.path.join(model, filename)
        os.rename(filepath, os.path.join(model, f"{bias}.pickle"))
        # also in csv dir
        filename = f"{bias}-sample.csv"
        filepath = os.path.join(f"{model}/csv", filename)
        os.rename(filepath, os.path.join(f"{model}/csv", f"{bias}.csv"))
```

```python
# rename negative_wording to allow_forbid
for model in models:
    exp_names = ["", "-key_typo", "-letter_swap", "-middle_random", "-steer2"]
    for exp in exp_names:
        filename = f"negative_wording{exp}.pickle"
        filepath = os.path.join(model, filename)
        if os.path.isfile(filepath):
            os.rename(filepath, os.path.join(model, f"allow_forbid{exp}.pickle"))
        # also in csv dir
        filename = f"negative_wording{exp}.csv"
        filepath = os.path.join(f"{model}/csv", filename)
        if os.path.isfile(filepath):
            os.rename(filepath, os.path.join(f"{model}/csv", f"allow_forbid{exp}.csv"))
```

```python
biases = ['acquiescence', 'allow_forbid', 'odd_even', 'opinion_float', 'response_order']
# combine diff pickle files
for model in models:
    print(model)
    exp_names = ["", "-key_typo", "-letter_swap", "-middle_random"]
    for bias in biases:
        for exp in exp_names:
            file1 = os.path.join(model, f"{bias}-50{exp}.pickle")
            file2 = os.path.join(model, f"{bias}-50diff{exp}.pickle")
            print(file1, file2)
            if os.path.isfile(file1) and os.path.isfile(file2):
                print('hi')
                file1_df = pd.read_pickle(file1)
                file2_df = pd.read_pickle(file2)
                combined_df = pd.concat([file1_df, file2_df], ignore_index=True)
                combined_df.to_pickle(f"{model}/{bias}{exp}.pickle")
            # also csv files
            file1 = os.path.join(f"{model}/csv", f"{bias}-50{exp}.csv")
            file2 = os.path.join(f"{model}/csv", f"{bias}-50diff{exp}.csv")
            if os.path.isfile(file1) and os.path.isfile(file2):
                file1_df = pd.read_csv(file1)
                file2_df = pd.read_csv(file2)
                combined_df = pd.concat([file1_df, file2_df], ignore_index=True)
                combined_df.to_csv(f"{model}/csv/{bias}{exp}.csv", index=False)
    
```
